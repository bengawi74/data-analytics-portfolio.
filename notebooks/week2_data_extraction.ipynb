```python
import pandas as pd
print("Notebook ready ‚úÖ")
print("Pandas:", pd.__version__)
```

```python
df = pd.read_csv("sample_data.csv")
df
```

```python
df["stock_value"] = df["quantity"] * df["price"]
df
```

```python
total_value = df["stock_value"].sum()
avg_price = df["price"].mean()
avg_quantity = df["quantity"].mean()

print("üí∞ Total Inventory Value:", total_value)
print("üè∑Ô∏è Average Price per Product:", avg_price)
print("üì¶ Average Quantity per Product:", avg_quantity)
```

```python
import matplotlib.pyplot as plt

# Create a bar chart
plt.figure(figsize=(6,4))
plt.bar(df["name"], df["stock_value"], color="skyblue", edgecolor="black")

# Add labels and title
plt.title("Stock Value by Product", fontsize=14)
plt.xlabel("Product Name")
plt.ylabel("Stock Value ($)")
plt.grid(axis="y", linestyle="--", alpha=0.6)

# Show the plot
plt.show()
```

```python
df_json = pd.read_json("sample_data.json")
df_json
```

```python
# Combine both datasets
combined_df = pd.concat([df, df_json], ignore_index=True)

# Calculate stock value for all rows
combined_df["stock_value"] = combined_df["quantity"] * combined_df["price"]

combined_df
```

```python
total_value_all = combined_df["stock_value"].sum()
avg_price_all = combined_df["price"].mean()
avg_quantity_all = combined_df["quantity"].mean()

print("üí∞ Total Inventory Value (All Sources):", total_value_all)
print("üè∑Ô∏è Average Price per Product:", avg_price_all)
print("üì¶ Average Quantity per Product:", avg_quantity_all)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(combined_df["name"], combined_df["stock_value"], color="lightgreen", edgecolor="black")

plt.title("Total Stock Value by Product (All Sources)", fontsize=14)
plt.xlabel("Product Name")
plt.ylabel("Stock Value ($)")
plt.grid(axis="y", linestyle="--", alpha=0.6)

# Display value labels on each bar
for i, v in enumerate(combined_df["stock_value"]):
    plt.text(i, v + 5, str(v), ha='center', fontsize=10)

plt.show()
```
